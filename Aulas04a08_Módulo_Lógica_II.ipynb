Criar uma função que extraia uma lista de nomes do dicionário recebido por parâmetro:

````python
def getNomes(lista):
  lista_nomes = []
  for aluno in lista:
    lista_nomes.append(aluno["Nome"] + " " + aluno["Sobrenome"])
  return lista_nomes

print("Segue a lista de nomes solicitada:", getNomes(lista_alunos))
````

Percorrendo dicionários usando funções de alta ordem (passando funções por parâmetro):

````python
def getNome(aluno):
  return aluno['Nome'] + ' ' + aluno["Sobrenome"]

# Esta é a implementação de como o map funciona por dentro (mais ou menos)
def mapNomes(getNome, lista):
  lista_nomes = []
  for aluno in lista:
    lista_nomes.append(getNome(aluno))
  return lista_nomes

print(mapNomes(getNome, lista_alunos))
print(list(map(getNome, lista_alunos)))
````

Agora um exemplo de uso do reduce: Como podemos obter a média dos CRs dos alunos?

````python
# Forma de fazer sem lambda function
def acumuladora(soma, aluno):
  cr_atual = aluno["CR"]
  print(f"Soma atual: {soma} - CR Atual {cr_atual}")
  return soma + cr_atual

# Obtendo a média dividindo a soma pela quantidade de alunos (len da lista)
print("Média:", reduce(acumuladora, lista_alunos, 0)/len(lista_alunos))
````

````python
# Forma de fazer com lambda function
soma_CRs = reduce(lambda soma, aluno: soma + aluno["CR"], lista_alunos, 0)

# Obtendo a média dividindo a soma pela quantidade de alunos (len da lista)
print("Média:", soma_CRs/len(lista_alunos))
print("Média:", soma_CRs / len(lista_alunos))
````

# Expressões geradoras para tuplas

Também é possível utilizar expressão geradoras para construir tuplas. Sendo assim, crie a função `getQuadrado()` que recebe uma tupla de elementos numéricos, e retorna uma tupla com o quadrado de cada elemento da tupla original.

````python
def getQuadrado(tupla):
  return tuple(elem**2 for elem in tupla)

getQuadrado((3, 5, 7, 10))
````

# Filtrando elementos por funções lambda

Sabendo disso, crie uma função filtraElementos() que recebe uma lista e utiliza função lambda para filtrar os elementos maiores que 10, ou seja, a função deve retornar uma lista apenas com estes elementos maiores que 10.

OBS: em um cenário real, a função filtraElementos() seria utilizada para outras funcionalidades também além da utilização da lambda, de forma a melhorar o determinismo do código.



````python
# Sem uso de lambda function
def maioresQue10(elem):
  return elem > 10

def filtraElementos(lista):
  return list(filter(maioresQue10, lista))

filtraElementos([2, 4, 15, 6, 11, 34])
````

````python
# Usando lambda function
def filtraElementos(lista):
  return list(filter(lambda elem: elem > 10, lista))

filtraElementos([2, 4, 15, 6, 11, 34])
````

# Compreensão de dicionários

Sendo assim, crie uma função mediaAlunosParaDicionario() que receba uma lista de listas, em que o primeiro elemento é uma lista com os nomes dos alunos; e o segundo elemento é uma lista com suas respectivas médias. Utilizando compreensão de dicionários, armazene estes dados no dicionário de forma que cada aluno seja a chave e sua média seja o valor.

OBS: caso a nota não esteja numérica, é necessário tratar os valores para tipos numéricos!



````python
def mediaAlunosParaDicionario(lista):
  lista_nomes = lista[0]
  lista_medias = lista[1]
  return { lista_nomes[i]: float(lista_medias[i]) for i in range(len(lista_medias)) }

lista = [
    ["Iago", "Ruan", "Wilza", "Márcia", "Orlando"],
    [4.3, 9.8, 9.3, 9.5, "9.4"]
]
mediaAlunosParaDicionario(lista)
````

Criar uma função que retorne o valor máximo ou mínimo de uma lista, determinado de acordo com um parâmetro (min)

````python
def getMinMax(num_list = [], get_min = True):
  try:
    num_min_max = num_list[0]
    for i in range(1, len(num_list)):
      if get_min:
        if num_list[i] < num_min_max:
          num_min_max = num_list[i]
      else:
        if num_list[i] > num_min_max:
          num_min_max = num_list[i]
    return num_min_max
  except IndexError:
    return "Lista vazia!"

getMinMax([12, 54, 23, 2, 56, 27], get_min = False)
getMinMax()
````

Exemplo de classe para os curiosos:

````python
class Quadrado():
  def __init__(self, x, y, lado):
    self.x = x
    self.y = y
    self.lado = lado

  def area(self):
    return self.lado ** 2

  def __gt__(self, q_outro):
    return self.area() > q_outro.area()

q1 = Quadrado(20, 30, 100)
q2 = Quadrado(40, 50, 120)

print(q1.x, q1.y, q1.lado)
print(q1.area())
print(q2.area())

if q2 > q1:
  print("O q2 é maior que o q1")
else:
  print("O q1 é maior que o q2")
````

````python
from functools import reduce
import json

# Ler uma lista de objetos de um JSON (100XP)
with open("bebidas.json", "r") as arq:
  lista_bebidas = json.loads(arq.read())

# Realizar um mapeamento, um filtro e uma redução (utilizar lambda function) (150XP);
# Obter uma lista com os nomes das bebidas
print("Nomes das bebidas:\n", list(map(lambda b: b["nome"], lista_bebidas)))

bebidas_em_estoque = list(filter(lambda b: b["em_estoque"], lista_bebidas))
print("Existem no estoque:\n", list(map(lambda b: b["nome"], bebidas_em_estoque)))

total_bebidas = len(lista_bebidas)
preco_medio = reduce(lambda soma, b: soma + b["preço"], lista_bebidas, 0)/total_bebidas
print(f"O preço médio das bebidas é: {preco_medio:.2f}")
````

Operação 1: LISTAGEM DE BEBIDAS

````python
def mostra_todas_bebidas(lista_bebidas):
  print("Segue a lista de todas as bebidas:")
  for bebida in lista_bebidas:
    print("-----------------------------")
    print("Nome:", bebida["nome"])
    print("Quantidade:", bebida["quantidade"])
    print("Preço:", bebida["preço"])
    print("Marca:", bebida["marca"])
    print("Em estoque:", bebida["em_estoque"])
    print("-----------------------------")

mostra_todas_bebidas(lista_bebidas)
````

````python
def atualiza_JSON(lista_bebidas):
  with open("bebidas.json", "w") as arq:
    arq.write(json.dumps(lista_bebidas))

def adicionar_bebida(lista_bebidas, bebida):
  lista_bebidas.append(bebida)
  atualiza_JSON(lista_bebidas)

adicionar_bebida(lista_bebidas, {
    "nome": "Cerveja maneira",
    "preço": 9.90,
    "quantidade": 14,
    "marca": "Heineken",
    "em_estoque": True,
})
````
